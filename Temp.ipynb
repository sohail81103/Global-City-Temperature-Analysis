```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('city_temperature.csv', low_memory=False)
df['date'] = pd.to_datetime(df[['Year', 'Month', 'Day']], errors='coerce')
df = df.dropna(subset=['date'])
city_df = df[df['City'] == 'Algiers'].copy()
city_df = city_df.sort_values('date')
```

```python
#### Temperature vs Time ####
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='date', y='AvgTemperature', data=city_df, color='darkorange')
plt.title('Average Daily Temperature in Algiers')
plt.xlabel('Date')
plt.ylabel('Avg Temperature (°F)')
plt.show()
city_df['month'] = city_df['date'].dt.month
```

```python
#### Monthly Avg Temperature ####
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='month', y='AvgTemperature', data=city_df, hue='month', palette='Spectral', legend=False)
plt.title('Monthly Temperature Distribution in Algiers')
plt.xlabel('Month')
plt.ylabel('Avg Temperature (°F)')
plt.show()
```

```python
#### Compare Multiple Cities ####
```

```python
cities = ['Algiers', 'Cairo', 'Nairobi']
multi_df = df[df['City'].isin(cities)].copy()
multi_df['date'] = pd.to_datetime(multi_df[['Year', 'Month', 'Day']])
plt.figure(figsize=(14,6))
sns.lineplot(data=multi_df, x='date', y='AvgTemperature', hue='City')
plt.title('Temperature Comparison: Algiers vs Cairo vs Nairobi')
plt.ylabel('Avg Temperature (°F)')
plt.show()
```
